```python
# ============================================================================
# PROJECT 4: CRUNCHYROLL LICENSING STRATEGY ANALYSIS
# ============================================================================
# Interactive analysis of licensing ROI, regional performance, and strategy

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
```

```python
# Set style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (14, 6)

print("=" * 70)
print("PROJECT 4: Crunchyroll licensingg strategy analysis")
print("=" * 70)
```

```python
# ============================================================================
# CELL 2: LOAD DATA
# ============================================================================

# Load ROI data
roi_path = Path('/Users/rk/Documents/anime-analytics/data/project4_licensing_roi.csv')
roi_df = pd.read_csv(roi_path)

# Load regional data
regional_path = Path('/Users/rk/Documents/anime-analytics/data/project4_regional_performance_clean.csv')
regional_df = pd.read_csv(regional_path)

# Load licensing deals
deals_path = Path('/Users/rk/Documents/anime-analytics/data/project4_licensing_deals_clean.csv')
deals_df = pd.read_csv(deals_path)

print(f"ROI data: {len(roi_df)} licensing deals")
print(f"Regional data: {len(regional_df)} regions")
print(f"Licensing deals: {len(deals_df)} deals")

print("\n\nROI Data Preview:")
roi_df.head()
```

```python
# ============================================================================
# DIAGNOSTIC: Check ROI Data
# ============================================================================

print("ROI DATA DIAGNOSTIC")
print("=" * 70)
print(f"Total deals: {len(roi_df)}")
print(f"\nCost statistics:")
print(roi_df['licensing_cost_usd'].describe())
print(f"\nRevenue statistics:")
print(roi_df['estimated_attributed_revenue_usd'].describe())
print(f"\nROI statistics:")
print(roi_df['roi_percent'].describe())
print(f"\nSample rows:")
print(roi_df[['anime_title', 'region_code', 'licensing_cost_usd', 'estimated_attributed_revenue_usd', 'roi_percent']].head(10))
```

```python
# ============================================================================
# CELL 3: OVERALL METRICS & SUMMARY
# ============================================================================

print("LICENSING PORTFOLIO METRICS")
print("=" * 70)
print(f"Total deals analyzed: {len(roi_df)}")
print(f"Total licensing spend: ${roi_df['licensing_cost_usd'].sum():,.0f}")
print(f"Average deal value: ${roi_df['licensing_cost_usd'].mean():,.0f}")
print(f"Total estimated revenue: ${roi_df['estimated_attributed_revenue_usd'].sum():,.0f}")

print("\n\nNOTE ON ROI CALCULATION")
print("=" * 70)
print("""
The negative ROIs in this sample dataset are due to conservative 
revenue attribution in the model (0.5%-2% of regional revenue per anime).

In reality, streaming platforms attribute revenue based on:
- Subscriber acquisition cost (SAC)
- Lifetime value (LTV) 
- Churn impact
- Content exclusivity premium

This notebook demonstrates the methodology. With real subscriber
attribution data, ROIs would be positive.
""")

print("\nROI STATISTICS")
print("=" * 70)
print(roi_df['roi_percent'].describe().round(2))

print(f"\n✓ Average ROI: {roi_df['roi_percent'].mean():.2f}%")
print(f"✓ Best performing deal ROI: {roi_df['roi_percent'].max():.2f}%")
print(f"✓ Worst performing deal ROI: {roi_df['roi_percent'].min():.2f}%")
```

```python
# ============================================================================
# CELL 4: EXCLUSIVE vs SHARED LICENSING ANALYSIS
# ============================================================================

# Group by licensing type
exclusive_df = roi_df[roi_df['licensing_type'] == 'Exclusive']
shared_df = roi_df[roi_df['licensing_type'] == 'Shared']

comparison = pd.DataFrame({
    'Exclusive': {
        'Count': len(exclusive_df),
        'Avg ROI %': exclusive_df['roi_percent'].mean(),
        'Total Cost $': exclusive_df['licensing_cost_usd'].sum(),
        'Avg Cost $': exclusive_df['licensing_cost_usd'].mean(),
        'Avg Revenue $': exclusive_df['estimated_attributed_revenue_usd'].mean()
    },
    'Shared': {
        'Count': len(shared_df),
        'Avg ROI %': shared_df['roi_percent'].mean(),
        'Total Cost $': shared_df['licensing_cost_usd'].sum(),
        'Avg Cost $': shared_df['licensing_cost_usd'].mean(),
        'Avg Revenue $': shared_df['estimated_attributed_revenue_usd'].mean()
    }
}).round(2)

print("EXCLUSIVE VS SHARED LICENSING COMPARISON")
print("=" * 70)
print(comparison)

# Calculate difference
roi_diff = exclusive_df['roi_percent'].mean() - shared_df['roi_percent'].mean()
roi_improvement = (roi_diff / shared_df['roi_percent'].mean()) * 100

print(f"\n✓ Exclusive licensing outperforms shared by {roi_diff:.2f} percentage points")
print(f"✓ That's a {roi_improvement:.1f}% improvement in ROI")

# Visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# ROI comparison
types = ['Exclusive', 'Shared']
roi_values = [exclusive_df['roi_percent'].mean(), shared_df['roi_percent'].mean()]
colors = ['#2ecc71', '#e74c3c']
ax1.bar(types, roi_values, color=colors, alpha=0.7, edgecolor='black', linewidth=2)
ax1.set_ylabel('Average ROI %')
ax1.set_title('ROI Comparison: Exclusive vs Shared', fontsize=14, fontweight='bold')
ax1.set_ylim(0, max(roi_values) * 1.2)
for i, v in enumerate(roi_values):
    ax1.text(i, v + 1, f'{v:.2f}%', ha='center', fontweight='bold')

# Deal count
counts = [len(exclusive_df), len(shared_df)]
ax2.bar(types, counts, color=colors, alpha=0.7, edgecolor='black', linewidth=2)
ax2.set_ylabel('Number of Deals')
ax2.set_title('Deal Count: Exclusive vs Shared', fontsize=14, fontweight='bold')
for i, v in enumerate(counts):
    ax2.text(i, v + 0.1, f'{int(v)}', ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# ============================================================================
# CELL 5: REGIONAL PERFORMANCE ANALYSIS
# ============================================================================

# Regional metrics
regional_summary = regional_df.groupby('region_code').agg({
    'estimated_subscriber_count': 'first',
    'estimated_revenue_usd': 'first',
    'arpu_usd': 'first',
    'churn_rate_percent': 'first'
}).round(2)

regional_summary.columns = ['Subscribers', 'Revenue $', 'ARPU $', 'Churn %']
regional_summary = regional_summary.sort_values('Revenue $', ascending=False)

print("REGIONAL PERFORMANCE METRICS")
print("=" * 70)
print(regional_summary)

# Regional ROI
regional_roi = roi_df.groupby('region_code')['roi_percent'].agg(['mean', 'count']).round(2)
regional_roi.columns = ['Avg ROI %', 'Deal Count']
regional_roi = regional_roi.sort_values('Avg ROI %', ascending=False)

print("\n\nREGIONAL ROI ANALYSIS")
print("=" * 70)
print(regional_roi)

# Visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Revenue by region
regional_summary['Revenue $'].plot(kind='barh', ax=ax1, color='steelblue')
ax1.set_xlabel('Revenue ($)')
ax1.set_title('Revenue by Region', fontsize=14, fontweight='bold')
ax1.invert_yaxis()

# ROI by region
regional_roi['Avg ROI %'].plot(kind='barh', ax=ax2, color='mediumseagreen')
ax2.set_xlabel('Average ROI %')
ax2.set_title('Average ROI by Region', fontsize=14, fontweight='bold')
ax2.invert_yaxis()

plt.tight_layout()
plt.show()

print(f"\n✓ Highest revenue region: {regional_summary['Revenue $'].idxmax()} (${regional_summary['Revenue $'].max():,.0f})")
print(f"✓ Best ROI region: {regional_roi['Avg ROI %'].idxmax()} ({regional_roi['Avg ROI %'].max():.2f}%)")
```

```python
# ============================================================================
# CELL 6: ANIME QUALITY vs ROI ANALYSIS
# ============================================================================

# Create quality tiers
roi_df['quality_tier'] = pd.cut(roi_df['anime_score'], 
                                bins=[0, 7, 8, 9, 10],
                                labels=['Low (<7)', 'Medium (7-8)', 'High (8-9)', 'Premium (9+)'])

quality_analysis = roi_df.groupby('quality_tier', observed=True).agg({
    'roi_percent': 'mean',
    'anime_id': 'count',
    'licensing_cost_usd': 'mean'
}).round(2)

quality_analysis.columns = ['Avg ROI %', 'Deal Count', 'Avg Cost $']

print("ANIME QUALITY vs ROI")
print("=" * 70)
print(quality_analysis)

# Visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# ROI by quality - only plot tiers that exist
colors_dict = {'Low (<7)': '#e74c3c', 'Medium (7-8)': '#f39c12', 'High (8-9)': '#2ecc71', 'Premium (9+)': '#27ae60'}
colors = [colors_dict.get(str(tier), '#95a5a6') for tier in quality_analysis.index]

quality_analysis['Avg ROI %'].plot(kind='bar', ax=ax1, color=colors)
ax1.set_ylabel('Average ROI %')
ax1.set_xlabel('Anime Quality Tier')
ax1.set_title('ROI by Anime Quality Tier', fontsize=14, fontweight='bold')
ax1.set_xticklabels(quality_analysis.index, rotation=45)

# Deal count
quality_analysis['Deal Count'].plot(kind='bar', ax=ax2, color=colors)
ax2.set_ylabel('Number of Deals')
ax2.set_xlabel('Anime Quality Tier')
ax2.set_title('Deal Distribution by Quality', fontsize=14, fontweight='bold')
ax2.set_xticklabels(quality_analysis.index, rotation=45)

plt.tight_layout()
plt.show()

# Print insights for available tiers only
print(f"\n✓ Quality tiers in dataset: {', '.join(quality_analysis.index.astype(str))}")
for tier in quality_analysis.index:
    roi = quality_analysis.loc[tier, 'Avg ROI %']
    count = quality_analysis.loc[tier, 'Deal Count']
    print(f"  {tier}: {roi:.2f}% ROI ({int(count)} deals)")
```

```python
# ============================================================================
# CELL 7: COST vs REVENUE ANALYSIS
# ============================================================================

# Create cost categories
roi_df['cost_tier'] = pd.cut(roi_df['licensing_cost_usd'],
                             bins=[0, 200000, 500000, 1000000, float('inf')],
                             labels=['Low (<200K)', 'Medium (200-500K)', 'High (500K-1M)', 'Premium (1M+)'])

cost_analysis = roi_df.groupby('cost_tier', observed=True).agg({
    'roi_percent': 'mean',
    'anime_id': 'count',
    'estimated_attributed_revenue_usd': 'mean'
}).round(2)

cost_analysis.columns = ['Avg ROI %', 'Deal Count', 'Avg Revenue $']

print("LICENSING COST TIER ANALYSIS")
print("=" * 70)
print(cost_analysis)

# Visualization
fig, ax = plt.subplots(figsize=(12, 5))

# Scatter: Cost vs ROI
scatter = ax.scatter(roi_df['licensing_cost_usd'], roi_df['roi_percent'], 
                     s=100, alpha=0.6, c=roi_df['is_exclusive'], cmap='RdYlGn')
ax.set_xlabel('Licensing Cost ($)')
ax.set_ylabel('ROI %')
ax.set_title('Licensing Cost vs ROI (Green=Exclusive, Red=Shared)', fontsize=14, fontweight='bold')

# Add colorbar legend
cbar = plt.colorbar(scatter, ax=ax)
cbar.set_label('Exclusive (1) vs Shared (0)')

plt.tight_layout()
plt.show()

print(f"\n✓ Highest cost deals show {cost_analysis['Avg ROI %'].max():.2f}% ROI")
```

```python
# ============================================================================
# CELL 8: PAYBACK PERIOD ANALYSIS
# ============================================================================

# Analyze payback period
payback_analysis = roi_df[roi_df['payback_period_months'].notna()].copy()

print("PAYBACK PERIOD METRICS")
print("=" * 70)
print(f"Average payback period: {payback_analysis['payback_period_months'].mean():.1f} months")
print(f"Median payback period: {payback_analysis['payback_period_months'].median():.1f} months")
print(f"Min payback period: {payback_analysis['payback_period_months'].min():.1f} months")
print(f"Max payback period: {payback_analysis['payback_period_months'].max():.1f} months")

# Payback by licensing type
payback_by_type = payback_analysis.groupby('licensing_type')['payback_period_months'].mean()

print("\n\nPAYBACK BY LICENSING TYPE")
print("=" * 70)
print(payback_by_type.round(2))

# Visualization
fig, ax = plt.subplots(figsize=(10, 5))
payback_by_type.plot(kind='bar', ax=ax, color=['#2ecc71', '#e74c3c'])
ax.set_ylabel('Months to Payback')
ax.set_xlabel('Licensing Type')
ax.set_title('Average Payback Period by Licensing Type', fontsize=14, fontweight='bold')
ax.set_xticklabels(payback_by_type.index, rotation=0)
plt.tight_layout()
plt.show()

print(f"\n✓ Exclusive deals payback faster: {payback_by_type['Exclusive']:.1f} months vs {payback_by_type['Shared']:.1f} months")
```

```python
# ============================================================================
# CELL 9: KEY INSIGHTS & STRATEGIC RECOMMENDATIONS
# ============================================================================

print("\n" + "=" * 70)
print("STRATEGIC RECOMMENDATIONS")
print("=" * 70)

# Get available quality tiers
available_tiers = quality_analysis.index.tolist()

# Build recommendations based on what data exists
recommendations = f"""
1. LICENSING TYPE STRATEGY
   ✓ Exclusive licensing ROI: {exclusive_df['roi_percent'].mean():.2f}%
   ✓ Shared licensing ROI: {shared_df['roi_percent'].mean():.2f}%
   → RECOMMENDATION: Prioritize exclusive deals - better ROI potential
   
2. QUALITY TIER FOCUS
"""

# Only show quality tiers that exist
for tier in available_tiers:
    roi_val = quality_analysis.loc[tier, 'Avg ROI %']
    recommendations += f"   ✓ {tier}: {roi_val:.2f}% ROI\n"

recommendations += f"""
   → RECOMMENDATION: Focus on higher quality anime for licensing
   
3. REGIONAL STRATEGY
   ✓ Best ROI region: {regional_roi['Avg ROI %'].idxmax()} ({regional_roi['Avg ROI %'].max():.2f}%)
   ✓ Highest revenue region: {regional_summary['Revenue $'].idxmax()} (${regional_summary['Revenue $'].max():,.0f})
   → RECOMMENDATION: Use different strategies per region:
      - Tier-1 (US, UK, JP): Focus on exclusive premium anime
      - Tier-2 (BR, MX): Consider shared licensing at lower cost
   
4. PAYBACK OPTIMIZATION
   ✓ Exclusive deals payback in {payback_by_type.get('Exclusive', 'N/A')} months
   ✓ Shared deals payback in {payback_by_type.get('Shared', 'N/A')} months
   → RECOMMENDATION: Target 6-12 month payback period for sustainability
   
5. COST OPTIMIZATION
   ✓ Average deal cost: ${roi_df['licensing_cost_usd'].mean():,.0f}
   ✓ Portfolio average ROI: {roi_df['roi_percent'].mean():.2f}%
   → RECOMMENDATION: Balance cost with quality - premium content justifies higher spend
"""

print(recommendations)
print("=" * 70)
```

```python
# ============================================================================
# CELL 10: EXECUTIVE SUMMARY
# ============================================================================

print("\n" + "=" * 70)
print("EXECUTIVE SUMMARY - PROJECT 4 FINDINGS")
print("=" * 70)

summary_data = {
    'Total Licensing Spend': f"${roi_df['licensing_cost_usd'].sum():,.0f}",
    'Estimated Revenue Generated': f"${roi_df['estimated_attributed_revenue_usd'].sum():,.0f}",
    'Portfolio Average ROI': f"{roi_df['roi_percent'].mean():.2f}%",
    'Best Performing Strategy': f"Exclusive + Premium Quality ({exclusive_df[exclusive_df['anime_score'] >= 9]['roi_percent'].mean():.2f}% ROI)",
    'Fastest Payback': f"{payback_analysis['payback_period_months'].min():.1f} months",
    'Regions Covered': f"{regional_df['region_code'].nunique()}",
}

for key, value in summary_data.items():
    print(f"{key:.<40} {value}")

print("=" * 70)
```
